```python
voting_series= pd.Series(df['Voting'])
def convert_voting(voting):
    # Remove parentheses and 'K'
    voting = voting.replace('(', '').replace(')', '').replace('K', '')
    
    # Convert the value to float first, then multiply by 1000 if needed
    try:
        if '.' in voting:  # Handle the case like '1.1K'
            return int(float(voting) * 1000)  # Convert to integer
        else:  # Handle the case like '53'
            return int(voting) * 1000  # Multiply by 1000 for 'K' values
    except ValueError:
        # If the conversion fails, return NaN or 0, depending on your preference
        return None  # You can also return np.nan or 0 here

# Apply the conversion to the Series
voting_series = voting_series.apply(convert_voting)

# Show the updated Series
print(voting_series)
```

```python
df['Votes'] =voting_series
df.drop('Voting',axis=1, inplace= True)
df

```

```python
import re

def duration_to_minutes(duration):
    # Extract hours (if any)
    h_match = re.search(r'(\d+)h', duration)
    # Extract minutes (if any)
    m_match = re.search(r'(\d+)m', duration)
    
    hours = int(h_match.group(1)) if h_match else 0
    minutes = int(m_match.group(1)) if m_match else 0
    
    return hours * 60 + minutes
df['Duration_min'] = df['Duration'].apply(duration_to_minutes)
df
```
